```python
def make_plot(GPs):
    mu = np.arange(mu_min, mu_max, mu_step)
    nlps = np.asarray([G.neg_log_p().data.numpy()[0] for G in GPs])
    good_indices = np.asarray([i for i in range(len(mu)) if i not in np.where(np.isnan(nlps))[0]
                              and nlps[i] < -1400])
    fig, ax = plt.subplots(figsize=(8,6))
    plt.plot(mu[good_indices], nlps[good_indices], linewidth=3, color='b')
    plt.xlabel('Hypothesized signal strength', fontsize=16)
    plt.ylabel('Negative Log Posterior', fontsize=16)
    plt.show()
```
